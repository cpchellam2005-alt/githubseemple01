```python
import pandas as pd

# Employee Dataset
data = {
    'Employee': [1, 2, 3, 4, 5, 6],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'HR', 'Sales'],
    'Salary': [30000, 50000, 45000, 35000, 32000, 40000],
    'Experience': [2, 5, 4, 3, 1, 6],
    'Attrition': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}


# Create Data Frame
df = pd.DataFrame(data)


# Display Dataset
print("Employee Dataset")
print(df)


# Display Column Meanings
print("\nDataset Column Descriptions")
print("Employee_ID : Unique ID assigned to each employee")
print("Department  : Employee's working department")
print("Salary      : Employee's monthly salary")
print("Experience  : Number of years worked")
print("Attrition   : Yes = Left company, No = Still working")
```

```python
import matplotlib.pyplot as plt

# Pie chart for Department distribution
department_counts = df['Department'].value_counts()

fig1, ax1 = plt.subplots()
ax1.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
ax1.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Employees by Department')
plt.show()
```
